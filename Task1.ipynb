Prompt Given to AI

Write 3 Python functions based on Formula 1 motorsport that are clean and reusable:

1, A lap time analyzer that takes a driver name and list of lap times, returns best, worst, average, and consistency status.

2, A fuel strategy calculator that takes total laps, fuel used per lap, and tank capacity, returns total fuel needed, number of pit stops, and laps per stint.

3, A championship points calculator that takes a list of drivers and their finishing positions, maps them to F1 points, and returns a sorted leaderboard.

Each function should print results AND return a dictionary so it can be reused in larger programs.

```python
##Function 1 — Lap Time Analyzer 🏁##

def analyze_lap_times(driver, lap_times):
    best = min(lap_times)
    worst = max(lap_times)
    avg = sum(lap_times) / len(lap_times)
    status = "Consistent" if (worst - best) < 1.5 else "Inconsistent"

    print(f"Driver  : {driver}")
    print(f"Best    : {best}s")
    print(f"Worst   : {worst}s")
    print(f"Average : {avg:.2f}s")
    print(f"Status  : {status}")

    return {"driver": driver, "best": best, "worst": worst,
            "average": round(avg, 2), "status": status}

result = analyze_lap_times("Verstappen", [83.4, 83.1, 83.8, 83.2, 83.5])
print(result)
```

```python
##Function 2 — Fuel Pit Stop Calculator ⛽##

def pit_stops_needed(total_laps, fuel_per_lap, tank_capacity):
    total_fuel = round(total_laps * fuel_per_lap, 2)
    pit_stops = int(total_fuel // tank_capacity)
    laps_per_stint = int(tank_capacity // fuel_per_lap)

    print(f"Total Fuel Needed : {total_fuel}kg")
    print(f"Pit Stops         : {pit_stops}")
    print(f"Laps per Stint    : {laps_per_stint}")

    return {"total_fuel_kg": total_fuel,
            "pit_stops": pit_stops,
            "laps_per_stint": laps_per_stint}

result = pit_stops_needed(57, 1.8, 30.0)
print(result)
```

```python
##Function 3 — Championship Points Leaderboard 🏆##

def championship_standings(drivers, positions):
    points_map = {1: 25, 2: 18, 3: 15, 4: 12, 5: 10,
                  6: 8,  7: 6,  8: 4,  9: 2,  10: 1}

    scores = {driver: points_map.get(pos, 0)
              for driver, pos in zip(drivers, positions)}

    leaderboard = sorted(scores.items(), key=lambda x: x[1], reverse=True)

    print("Championship Standings:")
    for rank, (driver, pts) in enumerate(leaderboard, 1):
        print(f"  P{rank}  {driver:<15} {pts} pts")

    return [{"rank": i+1, "driver": d, "points": p}
            for i, (d, p) in enumerate(leaderboard)]

drivers   = ["Verstappen", "Leclerc", "Hamilton", "Norris", "Alonso"]
positions = [1, 2, 3, 4, 5]

result = championship_standings(drivers, positions)
print(result)
```
